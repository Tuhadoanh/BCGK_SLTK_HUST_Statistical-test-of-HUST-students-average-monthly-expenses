```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# Đảm bảo Matplotlib hiển thị biểu đồ ngay trong notebook
%matplotlib inline

# 1. Đọc dữ liệu từ file CSV
filename = 'khao_sat_chi_tieu.csv'
df = pd.read_csv(filename)

# 2. Chọn cột dữ liệu 'ChiTieu' để phân tích
data = df['ChiTieu']

# 3. Hiển thị 5 dòng đầu tiên để kiểm tra
print(f"Đã đọc file '{filename}' thành công. 5 dòng đầu tiên:")
print(df.head())
```

```python
# Tính toán các giá trị mô tả
n = len(data)
x_bar = data.mean()
s = data.std()       # data.std() tự động là độ lệch chuẩn mẫu hiệu chỉnh (chia n-1)
median = data.median()

print("--- THỐNG KÊ MÔ TẢ (CHƯƠNG 3.1) ---")
print(f"Kích thước mẫu (n)            : {n}")
print(f"Trung bình mẫu (x_bar)         : {x_bar:.3f} triệu VNĐ")
print(f"Độ lệch chuẩn mẫu (s)        : {s:.3f} triệu VNĐ")
print(f"Trung vị (Median)             : {median:.3f} triệu VNĐ")
```

```python
print("--- ĐANG TẠO BIỂU ĐỒ (CHƯƠNG 3.1) ---")

# 1. Histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=8, edgecolor='black', alpha=0.7)
plt.title('Histogram Chi tiêu hàng tháng của Sinh viên HUST')
plt.xlabel('Chi tiêu (triệu VNĐ)')
plt.ylabel('Tần suất (Số lượng sinh viên)')
plt.axvline(x_bar, color='red', linestyle='--', label=f'Trung bình = {x_bar:.2f}')
plt.legend()
plt.savefig('images/histogram.png') # Lưu file
plt.show()
print("Đã lưu 'images/histogram.png'")

# 2. Boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(data, vert=False, patch_artist=True) # vert=False để vẽ nằm ngang
plt.title('Boxplot Chi tiêu hàng tháng')
plt.xlabel('Chi tiêu (triệu VNĐ)')
plt.savefig('images/boxplot.png') # Lưu file
plt.show()
print("Đã lưu 'images/boxplot.png'")
```

```python
alpha = 0.05
df = n - 1 # Bậc tự do (n - 1)

# Tính KTC 95%
ci_low, ci_high = stats.t.interval(confidence=1-alpha, 
                                 df=df, 
                                 loc=x_bar, 
                                 scale=stats.sem(data)) # stats.sem(data) là sai số chuẩn (s / sqrt(n))

print(f"--- KHOẢNG TIN CẬY 95% (CHƯƠNG 3.2) ---")
print(f"Bậc tự do (df) = {df}")
print(f"Khoảng tin cậy 95% cho mu là: ({ci_low:.3f}, {ci_high:.3f}) triệu VNĐ")
```

```python
mu_0 = 2.0  # Giá trị kiểm định từ giả thuyết H0

# Thực hiện t-test một mẫu, một phía (phía phải)
# alternative='greater' nghĩa là chúng ta đang kiểm định cho H1: mu > mu_0
t_statistic, p_value = stats.ttest_1samp(data, mu_0, alternative='greater')

print(f"--- KIỂM ĐỊNH GIẢ THUYẾT (CHƯƠNG 3.3) ---")
print(f"Giả thuyết H0: mu <= {mu_0}")
print(f"Giả thuyết H1: mu > {mu_0}")
print(f"Thống kê kiểm định (T_0)      : {t_statistic:.3f}")
print(f"P-value (cho kiểm định 1 phía): {p_value}") # In P-value ở dạng khoa học

# So sánh p-value với alpha
print("\n--- Kết luận ---")
if p_value < alpha:
    print(f"P-value ({p_value:.6f}) < Alpha ({alpha}).")
    print("Quyết định: Bác bỏ giả thuyết H0.")
else:
    print(f"P-value ({p_value:.6f}) >= Alpha ({alpha}).")
    print("Quyết định: Không đủ bằng chứng để bác bỏ H0.")
```
